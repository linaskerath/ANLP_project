```python
import pandas as pd
import rdata
```

```python
pd.set_option('display.max_columns', None)
```

```python
path_l2_trimmed = r"../data/version 1.1/primary data/eye tracking data/joint_data_l2_trimmed.rda"
path_l2_readrate = r"../data/version 1.1/primary data/eye tracking data/joint.l2.readrate.rda"
```

extract joint_data_l2_trimmed.rda

```python
l2_trimmed_parsed = rdata.parser.parse_file(path_l2_trimmed)
l2_trimmed_converted = rdata.conversion.convert(l2_trimmed_parsed)
df_trimmed = l2_trimmed_converted['joint.data']
df_trimmed
```

extract joint.l2.readrate.rda

```python
l2_parsed_readrate = rdata.parser.parse_file(path_l2_readrate)
l2_converted_readrate = rdata.conversion.convert(l2_parsed_readrate)
df_readrate = l2_converted_readrate['u']
df_readrate
```

extract joint.l2.readrate_sent.rda

```python
path_l2_readrate_sent = r"../data/version 1.1/primary data/eye tracking data/joint.l2.readrate_sent.rda"
l2_parsed_readrate_sent = rdata.parser.parse_file(path_l2_readrate_sent)
l2_converted_readrate_sent = rdata.conversion.convert(l2_parsed_readrate_sent)
```

```python
df_readrate_sent = l2_converted_readrate_sent['u2']
df_readrate_sent
```

```python
path_ind_diff = r"../data/version 1.1/primary data/individual differences data/joint.ind.diff.l2.rda"
```

```python
l2_parsed_readrate = rdata.parser.parse_file(path_ind_diff)
l2_converted_readrate = rdata.conversion.convert(l2_parsed_readrate)
```

```python
df_readrate = l2_converted_readrate['joint_id']
df_readrate
```
